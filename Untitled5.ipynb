Exercício 1 – Condicional Simples

Crie um programa que pergunte ao usuário se ele já montou sua árvore de Natal.

```python
def montar_arvore():
    resposta = input("Você já montou sua árvore de Natal? (sim/não): ")

    if resposta.lower() == "sim":
        print(" Que ótimo! O clima natalino já está tomando conta!")
    elif resposta.lower() == "não":
        print(" Ainda dá tempo! Monte a sua árvore hoje e aproveite o clima!")
    else:
        print(" Resposta inválida! Digite apenas 'sim' ou 'não'.")

montar_arvore()
print("\n")
```

Exercício 2 – Laço de Repetição

Peça ao usuário um número N e imprima uma contagem regressiva de N até 1, simulando um cronômetro para a chegada do Natal.

```python
def contagem_regressiva():
    N = int(input("Digite um número para a contagem regressiva: "))
    print("\nIniciando contagem:")

    for i in range(N, 0, -1):
        print(i)

    print(" Ho Ho Ho! O Natal chegou!")

contagem_regressiva()
print("\n")
```

Exercício 3 – Funções + Lista + Condicional

Crie uma função chamada verificar_presentes(lista_presentes) que recebe uma lista de presentes desejados para o Natal.

```python
def verificar_presentes(lista_presentes):
    print("\nAnalisando sua lista de presentes...")

    if len(lista_presentes) >= 3:
        print(" O Papai Noel terá muito trabalho este ano!")
    else:
        print(" Sua lista está muito curta!")

    return len(lista_presentes)


# EXTRA: função para criar automaticamente a lista
def criar_lista_presentes():
    lista = []
    while True:
        item = input("Digite um presente desejado (ou 'ok' para finalizar): ")
        if item.lower() == "ok":
            break
        lista.append(item)
    return lista


presentes_usuario = criar_lista_presentes()
total_presentes = verificar_presentes(presentes_usuario)
print("Total de presentes:", total_presentes)
print("\n")
```

Exercício 4 – Funções + Dicionário + Laços

Crie uma função contar_enfeites(enfeites) que recebe um dicionário onde:

```python
def contar_enfeites(enfeites):
    total = 0
    for chave, quantidade in enfeites.items():
        print(f"{chave}: {quantidade}")
        total += quantidade
    return total

# EXTRA: cadastrar enfeites
def cadastrar_enfeites():
    enfeites = {}
    while True:
        tipo = input("Tipo de enfeite (ou 'fim' para terminar): ").lower()
        if tipo == "fim":
            break
        qtd = int(input(f"Quantidade de {tipo}: "))
        enfeites[tipo] = qtd
    return enfeites

print("=== Cadastro de Enfeites ===")
enfeites_usuario = cadastrar_enfeites()
total = contar_enfeites(enfeites_usuario)
print(" Total de enfeites cadastrados:", total)
print("\n")
```

Exercício 5 – Função Completa (listas, dicionários, laço e condicional)

Crie um sistema simples para organizar uma ceia de Natal:

```python
def resumo_ceia(dados):
    # Quantidade de pessoas
    total_pessoas = len(dados)

    # Contar total de pratos
    total_pratos = sum(len(pratos) for pratos in dados.values())

    print("\n--- RESUMO DA CEIA ---")
    print(f"Pessoas contribuindo: {total_pessoas}")
    print(f"Total de pratos: {total_pratos}")

    # Condicional solicitada
    if total_pratos > 10:
        print("Essa ceia vai ser farta!")
    else:
        print("Ceia simples, mas especial!")


def cadastrar_ceia():
    ceia = {}

    print("=== Cadastro de pratos para a Ceia de Natal ===")
    print("Digite 'fim' para encerrar os nomes.\n")

    while True:
        pessoa = input("Nome da pessoa: ").strip()

        if pessoa.lower() == "fim":
            break

        pratos = []
        print(f"Digite os pratos que {pessoa} vai levar (digite 'ok' para finalizar):")

        while True:
            prato = input("Prato: ").strip()

            if prato.lower() == "ok":
                break

            pratos.append(prato)

        ceia[pessoa] = pratos

    return ceia


# ---------------------- FIM DO PROGRAMA ----------------------
if __name__ == "__main__":
    dados_ceia = cadastrar_ceia()
    resumo_ceia(dados_ceia)
    print("\n=== FIM ===")
```
